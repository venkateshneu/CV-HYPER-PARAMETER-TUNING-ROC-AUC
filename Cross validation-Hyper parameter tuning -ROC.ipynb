```python
df=pd.read_csv("https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv")
```

```python
df
```

```python
X=df.iloc[:,:-1]
```

```python
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_prob=lr.predict_proba(X_test)
```

```python
y_scores=y_pred_prob[:,1]
y_scores.shape
```

```python
X_train.shape
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_test,y_scores)
```

```python
thresholds
```

```python
fpr
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_scores)
```

```python
import seaborn as sns 
np.argmax(tpr)
tpr
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=3)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")

# Adding threshold values as markers on the ROC curve
for i, threshold in enumerate(thresholds):
    plt.text(fpr[i], tpr[i], f'{threshold:.2f}', fontsize=9, ha='center', va='bottom', color='blue')

plt.show()
```

```python
df =pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df
```

```python
X=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
knn=KNeighborsRegressor()
```

```python
kf=KFold(n_splits=5,  shuffle=True, random_state=42)
```

```python
scores=cross_val_score(knn,X,y,cv=kf,scoring="r2")
```

```python
np.mean(scores)
```

```python
# GridsearchCV
```

```python
knn=KNeighborsRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= {"n_neighbors":[1,3,5,7,9],
    "weights":['uniform','distance'],
   "algorithm":['ball_tree','kd_tree','brute'],
    "p":[1,2]
}
```

```python
gcv=GridSearchCV(knn,param_grid=param_grid,cv=kf,scoring="r2",verbose=2, refit=True)
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
a=pd.DataFrame(gcv.cv_results_)
```

```python
a[['param_algorithm', 'param_n_neighbors', 'param_p', 'param_weights','mean_test_score']].sort_values("mean_test_score", ascending=False)
```
